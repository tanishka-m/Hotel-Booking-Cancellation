# Data Analysis and Visualization

```python
calculate_perc = df['is_canceled'].value_counts(normalize=True) #finding percentage of cancelled & not cancelled booking
print(calculate_perc)

plt.figure(figsize = (5,4))   #fix the size of plot
plt.title("Reservation Status Count")
plt.bar(['Not Cancelled','Cancelled'],df['is_canceled'].value_counts(),edgecolor = 'k', width = 0.7) #plot the figure
```

```python
#reservation status in different hotel 
plt.figure(figsize = (8,4))
plt.title("Reservation status in different hotel", size = 20)
plt.xlabel('hotel')
plt.ylabel('number of reservation')
ax1=sns.countplot(x='hotel', hue='is_canceled', data = df, palette = 'Blues')

# legend_labels,_ = ax1. get_legend_handles_labels()
# ax1.legend(bbox_to_anchor=(1,1))
plt.show()
```

```python
#filter df by included canceled details of resort hotels
resort_hotel = df[df['hotel'] == 'Resort Hotel']#resort hotel cancelled in number
resort_hotel['is_canceled'].value_counts(normalize = True)
```

```python
#filter df by included canceled details of city hotels
city_hotel = df[df['hotel'] == 'City Hotel']#resort hotel cancelled in number
city_hotel['is_canceled'].value_counts(normalize = True)
```

```python
#group reservation dates based on month 
resort_hotel = resort_hotel.groupby("reservation_status_date")[['adr']].mean()
city_hotel = city_hotel.groupby("reservation_status_date")[['adr']].mean()
```

```python
#average daily rate of resort and city hotel wrt to time duration
plt.figure(figsize = (20,8))
plt.title('Average Daily Rate in City Hotel and Resort Hotel', fontsize=30)
plt.plot(resort_hotel.index,resort_hotel['adr'], label = 'resort hotel')
plt.plot(city_hotel.index,city_hotel['adr'], label = 'city hotel')
plt.legend(fontsize=20)
plt.show()
```

```python
#cancelation & not canceled room per month of both hotels
df['month'] = df['reservation_status_date'].dt.month
plt.figure(figsize = (13,6))
ax1 = sns.countplot(x = 'month' , hue = 'is_canceled', data = df, palette = 'bright')
plt.title("Reservation status per month", size = 20)
plt.xlabel('month')
plt.ylabel('number of reservation')
legend_labels,_ = ax1. get_legend_handles_labels()
ax1.legend(bbox_to_anchor=(1,1))
ax1.legend(['not canceled','canceled'])
plt.show()
```

```python
#cancellation per month
plt.figure(figsize = (13,6))
sns.barplot(x='month',y='adr', data = df[df['is_canceled'] == 1].groupby('month')[['adr']].sum().reset_index(), palette = 'pastel')
plt.title("ADR per month", size = 20)
plt.xlabel('month')
plt.ylabel('ADR')
plt.show()
```

```python
#country wise cancellation,consider only top 10 countries for analysis
cancelled_data = df[df['is_canceled'] == 1]
top_10_countries = cancelled_data['country'].value_counts()[:10]
plt.figure(figsize = (6,6))
plt.title('Top 10 countries with reservation canceled')
plt.pie(top_10_countries, autopct = '%.2f', labels = top_10_countries.index)
plt.show()
```

```python
#Identifiening our client source
df['market_segment'].value_counts()
```

```python
#Identifiening our client source with percentage contribution
df['market_segment'].value_counts(normalize = True)
```

```python
#cancelled data
cancelled_data['market_segment'].value_counts(normalize = True)
```

```python
cancelled_df_adr = cancelled_data.groupby ('reservation_status_date') [['adr']].mean()
cancelled_df_adr.reset_index(inplace = True)
cancelled_df_adr.sort_values('reservation_status_date', inplace = True)

#same with non cancalled data
not_cancelled_data = df [df['is_canceled'] == 0]
not_cancelled_df_adr = not_cancelled_data.groupby('reservation_status_date') [['adr']].mean()
not_cancelled_df_adr.reset_index(inplace = True)
not_cancelled_df_adr.sort_values('reservation_status_date', inplace= True)

plt.figure(figsize = (20,6))
plt.title('Average Daily Rate',fontsize = 20)
plt.plot(not_cancelled_df_adr['reservation_status_date'], not_cancelled_df_adr['adr'], label = 'not cancelled')
plt.plot(cancelled_df_adr['reservation_status_date'], cancelled_df_adr['adr'], label = 'cancelled')
plt.legend(fontsize = 20)
plt.show()
```

```python
#filtering data for the time duration 2016 - 2017-09
cancelled_df_adr = cancelled_df_adr[(cancelled_df_adr['reservation_status_date'] > '2016') & (cancelled_df_adr['reservation_status_date'] < '2017-09')]
not_cancelled_df_adr = not_cancelled_df_adr[(not_cancelled_df_adr['reservation_status_date'] > '2016') & (not_cancelled_df_adr['reservation_status_date'] < '2017-09')]
```

```python
#on filtered data
plt.figure(figsize = (20,6))
plt.title('Average Daily Rate',fontsize = 20)
plt.plot(not_cancelled_df_adr['reservation_status_date'], not_cancelled_df_adr['adr'], label = 'not cancelled')
plt.plot(cancelled_df_adr['reservation_status_date'], cancelled_df_adr['adr'], label = 'cancelled')
plt.legend(fontsize = 20)
plt.show()
```
